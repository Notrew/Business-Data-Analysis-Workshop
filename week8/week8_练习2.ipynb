```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airbnb=pd.read_csv('C:\\Users\mac\Desktop\数据分析班\week8\w3_airbnb.csv')
```

```python
#查看数据
airbnb.info()
```

    发现有三个变量为object，需要处理

```python
airbnb.describe()
```

    age变量有异常值

```python
#数据处理
#去除age变量异常值，取18到80岁之间的数据
airbnb=airbnb[airbnb["age"]<=80]
airbnb=airbnb[airbnb["age"]>=18]
airbnb.describe()
```

```python
#创建date衍生变量
#将date_account_created转换为日期时间型
airbnb["date_account_created"]=pd.to_datetime(airbnb["date_account_created"])
```

```python
#提取年份生产新变量—创建账户至今的时间
airbnb["year_since_account_created"]=airbnb["date_account_created"].apply(lambda x:2019-x.year)
airbnb.describe()
```

```python
#创建新变量—第一次预定的时间
airbnb["date_first_booking"]=pd.to_datetime(airbnb["date_first_booking"])
airbnb["year_since_first_booking"]=airbnb["date_first_booking"].apply(lambda x: 2019-x.year)
airbnb.describe()
```

```python
#删除两个原日期变量
airbnb.drop(airbnb.select_dtypes(["datetime64"]),inplace=True,axis=1)
```

```python
#将gender生产哑变量
airbnb=pd.get_dummies(airbnb)
airbnb.describe()
```

```python
#建立聚类分析模型
#基于行为偏好（下单渠道）和个人信息
airbnb_5=airbnb[["age","web","moweb","ios","android"]]
```

```python
#标准化数据
from sklearn.preprocessing import scale
x=pd.DataFrame(scale(airbnb_5))
```

```python
#创建模型
from sklearn import cluster
model=cluster.KMeans(n_clusters=3,random_state=10)
model.fit(x)
```

```python
airbnb_5["cluster"]=model.labels_
airbnb_5.head(20)
```

```python
#绘制散点图查看分类效果
sns.scatterplot(x="age",y="ios",hue="cluster",data=airbnb_5);
```

```python
sns.scatterplot(x="web",y="ios",hue="cluster",data=airbnb_5);
```

```python
#查看age的分群效果
airbnb_5.groupby(['cluster'])['age'].describe()
```

```python
#查看ios的分群效果
airbnb_5.groupby(['cluster'])['ios'].describe()
```

```python
#计算轮廓系数
from sklearn import metrics
x_cluster=model.fit_predict(x)
score=metrics.silhouette_score(x,x_cluster)
print(score)
```

```python
#模型优化
airbnb_5new=airbnb[['year_since_first_booking','web','moweb','ios','android']]
xnew=pd.DataFrame(scale(airbnb_5new))
```

```python
modelnew=cluster.KMeans(n_clusters=5,random_state=10)
modelnew.fit(xnew)
```

```python
x_new_cluster=modelnew.fit_predict(xnew)
score_new=metrics.silhouette_score(xnew,x_new_cluster)
print(score_new)
```
